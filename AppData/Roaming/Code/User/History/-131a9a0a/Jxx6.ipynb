2. Array Operations

```python
import numpy as np # type: ignore

# Re-establishing the variable from the previous step
years_exp = np.array([1, 3, 5, 7, 10])

# --- Basic Operations ---

# Step 1: Element-wise addition
exp_plus_5 = years_exp + 5
print("Years + 5:", exp_plus_5)

# Step 2: Element-wise multiplication
exp_times_2 = years_exp * 2
print("Years * 2:", exp_times_2)

# Step 3: Dot product
sample1 = np.array([1, 2, 3])
sample2 = np.array([4, 5, 6])
dot_result = np.dot(sample1, sample2)
print("Dot Product:", dot_result)

# --- "Try It" Section: Subtraction, Division, Exp, and Log ---

# Experiment with subtraction and division
exp_minus_1 = years_exp - 1
exp_div_2 = years_exp / 2

# Apply np.exp() and np.log()
# Using sample1 for cleaner log outputs (log(0) is undefined)
exp_values = np.exp(sample1) 
log_values = np.log(sample1)

print("-" * 30)
print("Subtraction (Years - 1):", exp_minus_1)
print("Division (Years / 2):", exp_div_2)
print("Exponential (e^x on [1,2,3]):", exp_values)
print("Natural Log (ln on [1,2,3]):", log_values)
```

3. Indexing and Slicing

```python
import numpy as np # type: ignore

# Data setup from previous steps
years_exp = np.array([1, 3, 5, 7, 10])
salaries = np.array([
    [50, 60, 70], 
    [80, 90, 100]
])

# Step 1: Access individual element
print("First year of experience:", years_exp[0])

# Step 2: Slice arrays (First two elements of first row)
print("First two salaries:", salaries[0, :2])

# Step 3: Access all rows for a specific column (Second column)
print("Second column salaries:", salaries[:, 1])

# Step 4: Negative indexing
print("Last year of experience:", years_exp[-1])

# --- "Try It" Section ---

# Reverse arrays using slicing
reversed_years = years_exp[::-1]
print("\nReversed years_exp:", reversed_years)

# Slice 2D arrays for subgroups of salaries 
# Example: Extracting a 2x2 sub-grid (Rows 0-2, Columns 1-3)
salary_subgroup = salaries[0:2, 1:3]
print("Subgroup of salaries (2x2 slice):\n", salary_subgroup)
```

6. Statistical Operations

```python
import numpy as np # type: ignore

# Data setup from previous steps
years_exp = np.array([1, 3, 5, 7, 10])
salaries = np.array([
    [50, 60, 70], 
    [80, 90, 100]
])

# Step 1: Mean of years of experience
print("Mean experience:", np.mean(years_exp))

# Step 2: Standard deviation of experience
print("Std deviation of experience:", np.std(years_exp))

# Step 3: Max and Min salaries
print("Max salary:", np.max(salaries), "| Min salary:", np.min(salaries))

# Step 4: Sum of salaries
print("Sum of all salaries:", np.sum(salaries))

# --- "Try It" Section ---

# Compute median
median_exp = np.median(years_exp)

# Compute 25th and 75th percentiles for salaries
percentile_25 = np.percentile(salaries, 25)
percentile_75 = np.percentile(salaries, 75)

print("-" * 30)
print("Median experience:", median_exp)
print("25th Percentile of Salaries:", percentile_25)
print("75th Percentile of Salaries:", percentile_75)
```

7. NumPy Functions on Arrays

```python
import numpy as np # type: ignore

# Data setup from previous steps
salaries = np.array([
    [50, 60, 70], 
    [80, 90, 100]
])

# Step 1: Apply trigonometric functions
# Angles in radians: 0, 45 degrees (pi/4), and 90 degrees (pi/2)
angles = np.array([0, np.pi/4, np.pi/2])
print("Sine of angles:", np.sin(angles))
print("Cosine of angles:", np.cos(angles))

# Step 2: Apply function along rows (axis=1)
# This sums the columns for each individual row
salary_sums = np.apply_along_axis(np.sum, 1, salaries)
print("Sum of Salaries per person:", salary_sums)

# --- "Try It" Section ---

# 1. Use np.sqrt() and np.log10() for transformations
sqrt_salaries = np.sqrt(salaries)
log_salaries = np.log10(salaries)

# 2. Apply a custom function (e.g., adding a 5% tax) using apply_along_axis
def add_tax(row):
    return row * 0.95  # Returns 95% of the original salary

taxed_salaries = np.apply_along_axis(add_tax, 1, salaries)

print("-" * 30)
print("Square Root of Salaries:\n", sqrt_salaries)
print("\nSalaries after 5% Tax (Custom Function):\n", taxed_salaries)
```

10. Data Grouping and Aggregation

```python
import pandas as pd # type: ignore
import numpy as np # type: ignore

# Step 1: Load the dataset with a safety check
try:
    df_kaggle = pd.read_csv('survey_results_public.csv')
    print("Stack Overflow Survey loaded successfully.")
except FileNotFoundError:
    df_kaggle = pd.read_csv('sample_data.csv')
    print("Survey file not found. Using sample_data.csv as a placeholder.")

# Step 2: Select relevant columns safely
# We check if 'Country' exists; if not, we use columns from your sample_data.csv
if 'Country' in df_kaggle.columns:
    target_cols = ['Country', 'EdLevel', 'YearsCodePro', 'ConvertedComp']
else:
    target_cols = ['X', 'Y', 'Z'] # Fallback for your current sample_data.csv

df_subset = df_kaggle[target_cols]

# Step 3: Clean the data
df_clean = df_subset.dropna().copy()

# Step 4: Categorize experience into groups
# If working with the real survey, we categorize by years
if 'YearsCodePro' in df_clean.columns:
    df_clean['YearsCodePro'] = pd.to_numeric(df_clean['YearsCodePro'], errors='coerce')
    df_clean = df_clean.dropna()
    df_clean['ExperienceLevel'] = np.where(df_clean['YearsCodePro'] >= 10, 'Senior', 'Junior')
else:
    # Placeholder logic for sample_data (X represents years)
    df_clean['ExperienceLevel'] = np.where(df_clean['X'] >= 25, 'Senior', 'Junior')

# Step 5 & 6: Group and Reset Index
group_by_col = 'Country' if 'Country' in df_clean.columns else 'Z'
value_col = 'ConvertedComp' if 'ConvertedComp' in df_clean.columns else 'Y'

grouped_data = df_clean.groupby([group_by_col, 'ExperienceLevel'])[value_col].mean().reset_index()
print("\nFormatted Grouped Data:\n", grouped_data.head())

# --- "Try It" Section ---

# 1. Change grouping to EdLevel and compute Median/Std
if 'EdLevel' in df_clean.columns:
    ed_stats = df_clean.groupby('EdLevel')['ConvertedComp'].agg(['median', 'std']).reset_index()
    print("\nEducation Stats (Median & Std):\n", ed_stats)

# 2. Explore top 10 countries with highest average compensation
if 'Country' in df_clean.columns:
    top_10 = grouped_data.groupby('Country')[value_col].mean().sort_values(ascending=False).head(10)
    print("\nTop 10 Highest Paying Countries:\n", top_10)
```
