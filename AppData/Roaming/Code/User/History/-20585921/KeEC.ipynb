```python
import pandas as pd
import numpy as np

# 1. Load the dataset (Using the name seen in your VS Code tab)
file_path = 'Most Popular Programming Languages.csv'

try:
    df = pd.read_csv(file_path)
    print("✅ File loaded successfully!")
    
    # 2. Setup
    df['Month'] = pd.to_datetime(df['Month'])
    languages = sorted([col for col in df.columns if col != 'Month'])
    
    # 3. Your Assignment: ID 363 -> Index 3
    student_id_last_3 = 363 
    lang_index = student_id_last_3 % 10
    assigned_lang = languages[lang_index]
    print(f"✅ Assigned Language: {assigned_lang}")

    # 4. Growth Analysis (Task 1)
    data = df[['Month', assigned_lang]].copy()
    data['Growth_Rate'] = data[assigned_lang].pct_change() * 100
    data['Moving_Avg'] = data[assigned_lang].rolling(window=12).mean()
    data['Moving_STD'] = data[assigned_lang].rolling(window=12).std()

    # Classification
    conditions = [(data['Growth_Rate'] > 5), (data['Growth_Rate'] < -5)]
    choices = ['Growth phase', 'Decline phase']
    data['Phase'] = np.select(conditions, choices, default='Stable phase')

    # 5. Lifecycle Classification (Task 2)
    overall_mean = data['Growth_Rate'].mean()
    overall_std = data['Growth_Rate'].std()
    
    # Lifecycle Rules
    life_cond = [
        (data['Growth_Rate'] > 0) & (data['Growth_Rate'] < overall_mean),
        (data['Growth_Rate'] > overall_mean),
        (data['Growth_Rate'].abs() <= 1),
        (data['Growth_Rate'] < 0) & (data['Growth_Rate'] < (-1 * overall_std))
    ]
    life_choices = ['Introduction', 'Growth', 'Maturity', 'Decline']
    data['Lifecycle_Phase'] = np.select(life_cond, life_choices, default='Stable')

    # --- FINAL OUTPUTS ---
    print("\n--- DataFrame Table (Last 5 Months) ---")
    print(data.tail())

    print("\n--- Statistical Summary ---")
    print(data[assigned_lang].describe())

    print("\n--- Dominant Lifecycle Stage ---")
    print(data['Lifecycle_Phase'].value_counts())

except Exception as e:
    print(f"❌ Error: {e}")
```
