```python
import numpy as np
import pandas as pd

STUDENT_NAME = "Roxie Nacional"
STUDENT_ID_STR = "TUPM-25-0363"

# Convert ID string to integer to fix the TypeError
student_id_int = int(''.join(filter(str.isdigit, STUDENT_ID_STR)))

# Seed for consistency
np.random.seed(student_id_int)

# 1. Create the DataFrame
df = pd.DataFrame({
    'EdLevel': ['Bachelor', 'Master', 'PhD', 'Bachelor', 'Master'],
    'ConvertedComp': np.random.randint(40000, 150000, 5)
})

# 2. Multi-metric Aggregation
# .agg() allows us to run multiple functions at once
summary = df.groupby('EdLevel')['ConvertedComp'].agg(['mean', 'median', 'std']).reset_index()

print(f"--- Education vs Salary Summary for {STUDENT_NAME} ---")
print(summary)
```
