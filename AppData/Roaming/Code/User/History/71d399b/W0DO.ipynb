1. Writing to a File

```python
# Step 1: Import Path from pathlib
from pathlib import Path

# Step 2: Define a custom folder in your Documents directory
output_dir = Path.home() / "Documents" / "Nacional_Activity_5"
output_dir.mkdir(exist_ok=True) 

# Step 3: Define the file path
file_path = output_dir / "Act5_example.txt"

# Step 4: Write text to the file
with open(file_path, "w", encoding="utf-8") as file:
    file.write("Hello, Welcome to Python Programming!\n")
    file.write("File saved safely with pathlib.")
    file.write("Python makes file handling easy!")

# Step 5: Confirm that the file is saved
print(f"File saved to: {file_path.resolve()}")
```

2. Reading from a File

```python
# Updated Step 2: Reading line-by-line with filters and word counts
with open(file_path, "r", encoding="utf-8") as file:
    print("--- Try It Results ---")
    for line_number, line in enumerate(file, 1):
        # Clean the line of extra spaces/newlines
        clean_line = line.strip()
        
        # 1. Experiment with .split() to count words
        words = clean_line.split()
        word_count = len(words)
        
        # 2. Modify the loop to print ONLY lines containing "Python"
        if "Python" in clean_line:
            print(f"Line {line_number} (Words: {word_count}): {clean_line}")
```

3. Appending Data

```python
# Step 1: Open file in append mode for a single line
with open(file_path, "a", encoding="utf-8") as file:
    file.write("\nThis line was added!")

# Try It: Append multiple lines using a list and writelines()
more_lines = [
    "\nLine from a list: Learning file I/O is fun.",
    "\nLine from a list: Python is powerful for ECE students."
]
with open(file_path, "a", encoding="utf-8") as file:
    file.writelines(more_lines)

# Try It: Ask the user for input and append it
user_note = input("Enter a custom note to add to the file: ")
with open(file_path, "a", encoding="utf-8") as file:
    file.write(f"\nUser Note: {user_note}")

# Step 2: Confirm
print("Data appended successfully.")
```

4. Safe File Operations with Backup

```python
# Step 1: Import additional libraries
from datetime import datetime
import shutil

# Step 2: Use your custom folder
backup_dir = Path.home() / "Documents" / "Nacional_Activity_5"
backup_dir.mkdir(exist_ok=True)

# Step 3: Function with "Try It" (Surname and Append/Overwrite choice)
def write_with_backup(filename: str, content: str, mode="w"):
    file_path = backup_dir / filename
    
    # Create backup if file exists
    if file_path.exists():
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        # Try It: Added Surname to the backup name
        backup_name = f"{file_path.stem}_Nacional_backup_{timestamp}{file_path.suffix}"
        backup_path = file_path.with_name(backup_name)
        shutil.copy2(file_path, backup_path)
        print(f"Backup saved: {backup_path.name}")
    
    # Try It: User choice for mode is handled by the 'mode' parameter
    with open(file_path, mode, encoding="utf-8") as file:
        file.write(content)
    print(f"File saved: {file_path.name} (Mode: {mode})")

# Try It: New function to list backups for a specific file
def list_specific_backups(filename: str):
    print(f"\n--- Listing backups for: {filename} ---")
    stem = Path(filename).stem
    # Searches for files that start with the filename and contain 'backup'
    for backup in backup_dir.glob(f"{stem}*backup*"):
        print("-", backup.name)

# --- Run Demonstration ---
print("=== File Operations Demo ===")

# 1. Create original
write_with_backup("demo.txt", "This is the first version.")

# 2. Update (This triggers the backup)
write_with_backup("demo.txt", "\nThis is the second version.", mode="a") # 'a' for append

# 3. List only backups for demo.txt
list_specific_backups("demo.txt")
```

5. Menu-Driven File Manager

```python
# Step 1: Import required modules
from pathlib import Path
from datetime import datetime
import shutil

# Try It: Change working directory to Surname_Activity_5
backup_dir = Path.home() / "Documents" / "Nacional_Activity_5"
backup_dir.mkdir(exist_ok=True)

def file_manager():
    file_name = input("Enter filename (e.g., notes.txt, data.csv, or readme.md): ")
    file_path = backup_dir / file_name
    
    while True:
        print("\n--- MENU ---")
        print("1. Write to file")
        print("2. Append to file")
        print("3. Read file")
        print("4. Backup file (with Surname)")
        print("5. List all backup files") # Added 6th option logic
        print("6. Exit")
        
        choice = input("Choose an option (1-6): ")
        
        if choice == "1":
            content = input("Enter content to write:\n")
            with open(file_path, "w", encoding="utf-8") as f:
                f.write(content)
            print("File written successfully.")
            
        elif choice == "2":
            more = input("Enter content to append:\n")
            with open(file_path, "a", encoding="utf-8") as f:
                f.write("\n" + more)
            print("Content appended.")
            
        elif choice == "3":
            if file_path.exists():
                with open(file_path, "r", encoding="utf-8") as f:
                    print(f"\n--- Content of {file_name} ---\n", f.read())
            else:
                print("File not found.")
                
        elif choice == "4":
            if file_path.exists():
                timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
                # Try It: Modify backup name to include Surname
                backup_file = file_path.with_name(
                    f"{file_path.stem}_Nacional_backup_{timestamp}{file_path.suffix}"
                )
                shutil.copy2(file_path, backup_file)
                print(f"Backup created: {backup_file.name}")
            else:
                print("Cannot backup. File does not exist.")
        
        elif choice == "5":
            # Try It: 6th menu option to list all backup files
            print("\n--- List of All Backups ---")
            backups = list(backup_dir.glob("*backup*"))
            if backups:
                for b in backups:
                    print(f"- {b.name}")
            else:
                print("No backups found.")
                
        elif choice == "6":
            print("Exiting the file manager.")
            break
        else:
            print("Invalid choice. Please try again.")

# Run the tool
if __name__ == "__main__":
    file_manager()
```
