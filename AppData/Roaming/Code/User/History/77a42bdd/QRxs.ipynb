```python
# [CELL 1] STUDENT CONFIGURATION

LAST_NAME = "Nacional"
STUDENT_ID = "TUPM-25-0363"

SEED_DIGIT = int(STUDENT_ID[-1])
ID_SUM = sum(int(d) for d in STUDENT_ID if d.isdigit())
NAME_LENGTH = len(LAST_NAME)

print("Student:", LAST_NAME)
print("Seed Digit:", SEED_DIGIT)
print("Digit Sum:", ID_SUM)
print("Surname Length:", NAME_LENGTH)
```

```python
# [CELL 2] FUNCTIONAL ENCAPSULATION

def greet():
    print("=" * 40)
    print(f"System Initialized for: {LAST_NAME}")
    print("=" * 40)

def identity_code():
    code = SEED_DIGIT * NAME_LENGTH
    return code

greet()
print("Identity Code:", identity_code())
```

```python
# [CELL 3] LEGB VALIDATION

global_value = ID_SUM

def scope_test():
    global_value = SEED_DIGIT
    print("Inside function:", global_value)

scope_test()
print("Outside function:", global_value)
```

```python
# [CELL 4] PARAMETRIC PROCESSING

def user_summary(title, *scores, **info):
    print(f"=== {title} ===")

    total = sum(scores)
    average = total / len(scores)

    print("Scores:", scores)
    print("Total:", total)
    print("Average:", round(average, 2))

    for k,v in info.items():
        print(f"(k): (v)")

    return average

avg = user_summary(
    f"{LAST_NAME} Academic Report", 
    SEED_DIGIT * 10,
    ID_SUM % 100,
    NAME_LENGTH * 5,
    id=STUDENT_ID,
    surname=LAST_NAME
)
```

```python
# [CELL 5] RETURN MECHANISM

def compute_area(radius):
    return 3.1416 * radius**2

area = compute_area (SEED_DIGIT + 2)
print("Computed Area:", round(area, 2))
```

```python
# [CELL 6] SAFE CALCULATOR

def safe_calculator(x, y):
    try:        
        result = x/y
    except ZeroDivisionError:
        result = "UNDEFINED"
    return result

print("Calculation Result:", safe_calculator(ID_SUM, SEED_DIGIT))
```

```python
# [CELL 7] RECURSION

def factorial(n):
    if n <= 1:
        return 1
    return n * factorial(n-1)

print("Factorial Result:", factorial (SEED_DIGIT))
```

```python
# [CELL 8] STRING RECURSION

def reverse(text):
    if len(text) <= 1:
        return text
    return text[-1] + reverse(text[:-1])

print("Reversed Surname:", reverse(LAST_NAME.upper()))
```

```python
# [CELL 9] DECORATOR

def logger(func):
    def wrapper(*args, **kwargs):
        # Change this line: use DOUBLE underscores
        print("Executing:", func.__name__) 
        return func(*args, **kwargs)
    return wrapper

@logger
def multiply(x, y):
    return x * y

# Ensure SEED_DIGIT and NAME_LENGTH are defined in a previous cell!
print("Multiplication:", multiply(SEED_DIGIT, NAME_LENGTH))
```

```python
# [CELL 10] LAMBA

data = list(range (1, SEED_DIGIT + 10))
squared = list(map(lambda x: x**2, data))

print("Squared Data:", squared)
```

```python
# [CELL 11] GENERATOR

def even_squares (limit):
    for i in range(limit):
        if 1 % 2 == 0:
            yield 1**2

for value in even_squares (NAME_LENGTH + SEED_DIGIT):
    print(value)
```

```python
# Mission 2: Recursive Signal Shutdown
def signal_shutdown(power, count=0):
    if power <= 0:
        print("Signal strength: 0 - Shutdown Complete")
        return count
    
    print(f"Current signal strength: {power}")
    return signal_shutdown(power - 1, count + 1)
```

```python
# Mission 3: Streaming Media Analytics (Generator)
def media_analytics_gen(limit):
    for i in range(limit + 1):
        if i % 2 == 0:
            yield i ** 2
```

```python
# MISSION 2: RECURSIVE SHUTDOWN
def signal_shutdown(power, count=0):
    # Base Case: Stop when power hits 0
    if power <= 0:
        print(f"Signal Strength: 0% - Critical Shutdown Initiated.")
        return count
    
    print(f"Signal Strength: {power}% - Stable...")
    # Recursive Call: Decrement power, increment call count
    return signal_shutdown(power - 1, count + 1)

# Usage Example:
# initial_power = 3 + 12 = 15
# total_calls = signal_shutdown(15)
# print(f"Total Recursive Calls: {total_calls}")
```
